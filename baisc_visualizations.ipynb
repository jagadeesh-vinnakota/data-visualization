```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Baisc plot
plt.plot([2,4,6,8],[3,5,8,10])
plt.title("basic visualization")
plt.xlabel("x-data")
plt.ylabel("y-data")
plt.show()
```

```python
# Baisc dotted plot
plt.plot([2,4,6,8],[3,5,8,10], "go")
plt.title("basic visualization")
plt.xlabel("x-data")
plt.ylabel("y-data")
plt.show()
```

```python
# Mutiple data set visualization
plt.plot([2,4,6,8],[3,5,8,10], "go", [1,2,3,4], [1,4,9,16], "r^")
plt.title("basic visualization")
plt.xlabel("x-data")
plt.ylabel("y-data")
plt.show()
```

```python
# Mutiple plots in one figure
plt.subplot(1,3,1)
plt.plot([2,4,6,8],[3,5,8,10], "go")
plt.title("basic visual")
plt.xlabel("x-data")
plt.ylabel("y-data")
plt.subplot(1,3,3)
plt.plot([1,2,3,4], [1,4,9,16], "r^")
plt.title("basic visual")
plt.xlabel("x-data")
plt.ylabel("y-data")
plt.suptitle("Mutiplot")
plt.show()
```

```python
# Creating multiple plots using subplots method
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(8,8))
ax[0,0].plot([2,4,6,8],[3,5,8,10], "go")
ax[0,0].set_title("plot1")
ax[1,1].plot([1,2,3,4], [1,4,9,16], "r^")
ax[1,1].set_title("plot2")
plt.suptitle("Multi plots")
plt.show()
```

```python
#Bar Graph
divisions = ['A', 'B', 'C', 'D']
population_count = [200,140,400,300]
plt.bar(divisions, population_count, color='green')
plt.xlabel("Divisions")
plt.ylabel("Population counts")
plt.show()
```

```python
import numpy as np
# Horizontal stacked bar graphs
divisions = ['A', 'B', 'C', 'D', 'E']
male_count = [200,140,400,300,500]
female_count = [150,100,320,400, 450]

index = np.arange(5)
width = 0.30

plt.bar(index,male_count, width, label = 'Male count', color='green')
plt.bar(index+width, female_count, width, label = 'Female count', color = 'red')

plt.xlabel("Divisions")
plt.ylabel("Population counts")
plt.xticks(index+width/2, divisions)
plt.legend(loc = 'best')
plt.show()
```

```python
# Histograms

data = np.random.randn(1000)
plt.title("Histograms")
plt.xlabel("Random data")
plt.ylabel("frequency")
plt.hist(data, 15)
plt.show()
```

```python
# Scatter Plots
height = [160,170,180,165,156]
weight = [70,76,89,67,65]

plt.scatter(height, weight, color = 'red')
plt.xlabel("height")
plt.ylabel("weight")
plt.show()
```
